# Extracting and Preprocessing Text from diverse Sources

How to go about and work with Textual Data in multiple sourcefiles

### PDF Files

A very common Form of textual Data is PDF. Let's see how to go about working with it in Python

```python
import PyPDF2 as pdf
 
# Open the pdf file
file = open('./data/test2.pdf', 'rb')
# And generate the Reader object for the file
pdfReader = pdf.PdfFileReader(file)
```

```python
# Extract the Metainformation on the Document
pdfReader.documentInfo
```

```python
# We can get the number of pages
pdfReader.getNumPages()
```

```python
# We can read a page
pdfReader.getPage(0)
```

#### Extracting the Text from the PDF

```python
# Extract Text from PDF
page = pdfReader.getPage(3)
page_content = page.extractText()
```

```python
page_content[:1000]
```

#### Poor Quality 

We see that the content of the page has been picked up but in very poor quality, and it does not only contain layout information, but also additional values intermingled into the text. Let's see what we can do to clean this up
