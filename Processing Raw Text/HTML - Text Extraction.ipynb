# Loading Text from HTML

```python
# Basic imports
from pprint import pprint
import nltk
from bs4 import BeautifulSoup as bs
```

```python
# Import Text from the web
from urllib import request
url = "https://www.nytimes.com/2017/09/03/us/quillen-small-business-michigan.html?hp&action=click&pgtype=Homepage&clickSource=story-heading&module=first-column-region&region=top-news&WT.nav=top-news"
response = request.urlopen(url)
soup = bs(response, 'lxml')
```

```python
pprint(soup.prettify())
```

```python
header = soup.find_all('h1')
corpus = soup.find_all('p')
```

```python
text = [p.get_text() for p in corpus]
```
